```python
import pandas as pd
from sklearn.impute import SimpleImputer

# Convierte columnas a valores numéricos si es posible
numeric_columns = ['Serial Number', 'List Year', 'Assessed Value', 'Sale Amount', 'Sales Ratio', 'Location']
for column in numeric_columns:
    if column in df_final.columns:
        df_final[column] = pd.to_numeric(df_final[column], errors='coerce')

# Define estrategias de imputación para cada columna
imputation_strategies = {
    'Serial Number': 'median',
    'List Year': 'median',
    'Date Recorded': 'most_frequent',
    'Town': 'most_frequent',
    'Address': 'most_frequent',
    'Assessed Value': 'median',
    'Sale Amount': 'median',
    'Sales Ratio': 'median',
    'Property Type': 'most_frequent',
    'Residential Type': 'most_frequent',
    'Non Use Code': 'most_frequent',
    'Assessor Remarks': 'most_frequent',
    'OPM remarks': 'most_frequent',
    'Location': 'median'
}

# Imputación basada en las estrategias definidas
for column, strategy in imputation_strategies.items():
    if column in df_final.columns:
        # Crear el imputador según el tipo de columna
        if df_final[column].dtype in ['float64', 'int64']:  # Solo columnas numéricas
            imputer = SimpleImputer(strategy=strategy)
            df_final[[column]] = imputer.fit_transform(df_final[[column]])
        else:  # Columnas categóricas
            imputer = SimpleImputer(strategy='most_frequent')
            df_final[[column]] = imputer.fit_transform(df_final[[column]])
    else:
        print(f"Columna {column} no encontrada en el DataFrame.")

print("Imputación completada.")
```

```python
def missing_cols(df):
    total=0
    for col in df.columns:
        missing_vals=df[col].isnull().sum()
        total +=missing_vals
    if missing_vals != 0:
        print(f"{col} => {df[col].isnull().sum()}")
    if total==0:
        print ("No hay valores faltantes")

missing_cols(df_final)
```

```python
import matplotlib.pyplot as plt

# Filtrar los 10 primeros elementos en el DataFrame basado en la columna 'Property Type'# Contar la frecuencia de cada tipo de propiedad y seleccionar los 10 más frecuentes
top_10_property_types = df_final['Property Type'].value_counts().nlargest(10)

# Convertir los datos filtrados en listas
etiqueta = top_10_property_types.index.tolist()
counts = top_10_property_types.tolist()

fig, ax = plt.subplots(1, 1, figsize=(15, 6.9), dpi=100)

# Colores para las barras (puedes personalizarlos)
bar_colors = ['#f75e50'] + ['#DCDCDC'] * 9# Ejemplo de colores; ajusta según sea necesario# Crear el gráfico de barras horizontal con los datos filtrados
ax.barh(etiqueta, counts, color=bar_colors)

# ----------------------------------------------------------------------------------------------------------------------------## Formato del gráfico.

titulo = 'Grafico 1: Top 10 Tipos de Propiedad'
plt.suptitle(titulo,
             horizontalalignment='left',
             x=0.001,
             y=0.87,
             transform=fig.transFigure,
             fontsize=20,
             fontweight='bold',
             fontfamily='serif')

subtitulo = 'Cantidad de propiedades por tipo'
ax.set_title(subtitulo,
             ha='left',
             x=0.70,
             y=0.,
             transform=fig.transFigure,
             fontsize=12,
             fontweight='regular',
             fontfamily='serif')

etiqueta = 'Cantidad de propiedades'
ax.set_xlabel(etiqueta,
              ha='left',
              fontsize=10,
              fontweight='regular',
              fontfamily='serif')

# Ajustes del gráfico
ax.xaxis.set_label_coords(x=0, y=-0.06, transform=ax.transAxes)
plt.subplots_adjust(top=0.80)

# Guardar el gráfico como una imagen
plt.savefig('Graf1.png')

# Mostrar el gráfico
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Asegúrate de que la columna 'Sale Amount' esté en formato numérico
df_final['Sale Amount'] = pd.to_numeric(df_final['Sale Amount'], errors='coerce')

# Elimina filas con valores nulos en 'Sale Amount' y 'Property Type'
df_final = df_final.dropna(subset=['Sale Amount', 'Property Type'])

# Calcula el costo promedio por tipo de propiedad
costo_promedio = df_final.groupby('Property Type')['Sale Amount'].mean()

# Configura el gráfico
fig, ax = plt.subplots(figsize=(10, 6), dpi=100)

# Graficar el costo promedio
costo_promedio.plot(kind='bar', color='skyblue', ax=ax)

# Etiquetas y título
ax.set_xlabel('Tipo de Propiedad', fontsize=12)
ax.set_ylabel('Costo Promedio ($)', fontsize=12)
ax.set_title('Costo Promedio de Propiedades por Tipo', fontsize=14, fontweight='bold')

# Ajustar el formato del gráfico
plt.xticks(rotation=45)  # Rotar etiquetas del eje x si es necesario
plt.tight_layout()       # Ajustar el diseño para que se vea bien# Guardar el gráfico como una imagen
plt.savefig('Costo_Promedio_Por_Tipo.png')

# Mostrar el gráfico
plt.show()
```

```python
import pandas as pd

# Asegúrate de que las columnas 'Sale Amount' y 'Assessed Value' estén en formato numérico
df_final['Sale Amount'] = pd.to_numeric(df_final['Sale Amount'], errors='coerce')
df_final['Assessed Value'] = pd.to_numeric(df_final['Assessed Value'], errors='coerce')

# Elimina filas con valores nulos en las columnas clave
df_final = df_final.dropna(subset=['Sale Amount', 'Assessed Value'])

# Calcula el margen de ganancia para cada propiedad
df_final['Gain Margin (%)'] = ((df_final['Sale Amount'] - df_final['Assessed Value']) / df_final['Sale Amount']) * 100# Calcula el margen promedio de ganancia
margen_promedio = df_final['Gain Margin (%)'].mean()

print(f"Margen promedio de ganancia por propiedad: {margen_promedio:.2f}%")
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Asegúrate de que las columnas 'Sale Amount' y 'Assessed Value' estén en formato numérico
df_final['Sale Amount'] = pd.to_numeric(df_final['Sale Amount'], errors='coerce')
df_final['Assessed Value'] = pd.to_numeric(df_final['Assessed Value'], errors='coerce')

# Elimina filas con valores nulos en las columnas clave
df_final = df_final.dropna(subset=['Sale Amount', 'Assessed Value', 'Property Type'])

# Calcula el margen de ganancia para cada propiedad
df_final['Gain Margin (%)'] = ((df_final['Sale Amount'] - df_final['Assessed Value']) / df_final['Sale Amount']) * 100# Calcula el margen promedio de ganancia por tipo de propiedad
margen_promedio_por_tipo = df_final.groupby('Property Type')['Gain Margin (%)'].mean()

# Configura el gráfico
fig, ax = plt.subplots(figsize=(12, 7), dpi=100)

# Graficar el margen promedio de ganancia
margen_promedio_por_tipo.plot(kind='bar', color='lightcoral', ax=ax)

# Etiquetas y título
ax.set_xlabel('Tipo de Propiedad', fontsize=12)
ax.set_ylabel('Margen Promedio de Ganancia (%)', fontsize=12)
ax.set_title('Margen Promedio de Ganancia por Tipo de Propiedad', fontsize=14, fontweight='bold')

# Ajustar el formato del gráfico
plt.xticks(rotation=45)  # Rotar etiquetas del eje x si es necesario
plt.tight_layout()       # Ajustar el diseño para que se vea bien# Guardar el gráfico como una imagen
plt.savefig('Margen_Promedio_Ganancia_Por_Tipo.png')

# Mostrar el gráfico
plt.show()
```
