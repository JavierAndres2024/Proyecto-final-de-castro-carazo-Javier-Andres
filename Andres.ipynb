<h2 Style = "color:  #2790b0 ">Eliminar observaciones del Data Frame</h2>

```python
dfAutos.dropna(subset=['price'], axis=0, inplace=True)
dfAutos.reset_index(drop=True, inplace=True)
dfAutos
```

```python
def missing_cols(df):
    total=0
    for col in df.columns:
        missing_vals=df[col].isnull().sum()
        total +=missing_vals
    if missing_vals != 0:
        print(f"{col} => {df[col].isnull().sum()}")
    if total==0:
        print ("No hay valores faltantes")

missing_cols(dfAutos)
```

```python
dfg = dfAutos.groupby(['make']).count()
dfg = dfg.iloc[:,:1]
dfg = dfg.sort_values('symboling', ascending = False)
dfg.head(5)
```

```python
fig, ax = plt.subplots(1,1, figsize = (15,6.9), dpi = 100)

etiqueta = dfg.index.tolist()
counts = dfg['symboling'].tolist()
bar_colors = ['#f75e50','#DCDCDC','#DCDCDC','#DCDCDC','#DCDCDC','#DCDCDC','#DCDCDC','#DCDCDC','#DCDCDC','#DCDCDC','#DCDCDC',
              '#DCDCDC','#DCDCDC','#DCDCDC','#DCDCDC','#DCDCDC','#DCDCDC','#DCDCDC','#DCDCDC','#DCDCDC','#DCDCDC',
              '#7d7769']

ax.barh(etiqueta, counts, color=bar_colors)
#----------------------------------------------------------------------------------------------------------------------------#
#Formato del grafico.

titulo = 'Grafico 1: Inventario de Vehiculos'
plt.suptitle(titulo,
             horizontalalignment = 'left',
            x = 0.001,
            y = 0.87,
            transform = fig.transFigure,
            fontsize = 20,
            fontweight = 'bold',
            fontfamily = 'serif')

subtitulo = 'Cantidad de carros segun fabricante'
ax.set_title(subtitulo,
             ha = 'left',
            x = 0.001,
            y = 0.80,
            transform = fig.transFigure,
            fontsize = 12,
            fontweight = 'regular',
            fontfamily = 'serif')

etiqueta = 'Cantidad de vehiculos'
ax.set_xlabel(etiqueta,
             ha = 'left',
             fontsize = 10,
             fontweight = 'regular',
             fontfamily = 'serif')

ax.xaxis.set_label_coords(x=0, y= -0.06, transform=ax.transAxes)
ax.text(46.5, -0.409, "46")
plt.subplots_adjust(top=0.80)

plt.savefig('Graf1.png')

plt.show()
```
